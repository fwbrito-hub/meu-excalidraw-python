## 🧪 Lição 1: Testes Automatizados com Pytest

Testar manualmente cada rota toda vez que você muda uma vírgula é perda de tempo. Usaremos o `pytest` e o `httpx` para simular um cliente acessando nossa API.

````python
# Exemplo de um teste de rota (test_main.py)
from fastapi.testclient import TestClient
from app.main import app

client = TestClient(app)

def test_read_main():
    # 1. ACIONA A ROTA
    response = client.get("/")
    
    # 2. VALIDA O RESULTADO
    assert response.status_code == 200
    assert response.json() == {"status": "Operacional", "servidor": "Ligado", "banco": "SQLite Ativo"}

def test_salvar_projeto_incompleto():
    # Testando se a API recusa dados errados (Isso é qualidade!)
    response = client.post("/api/projetos/salvar", json={"nome": "Erro"})
    assert response.status_code == 422 # Erro de validação do Pydantic
````

## 📦 Lição 2: Docker - A Carga em Containers

O Docker resolve o problema clássico: *"Na minha máquina funciona!"*. 
Ele empacota o Python, as bibliotecas e o seu código em uma imagem isolada.

### O `Dockerfile` do Backend (Resumo):
```dockerfile
FROM python:3.11-slim
WORKDIR /app
COPY requirements.txt .
RUN pip install -r requirements.txt
COPY . .
CMD ["uvicorn", "app.main:app", "--host", "0.0.0.0", "--port", "8000"]
```

## 🎼 Lição 3: Docker Compose - O Maestro

Para rodar o Backend (Python) e o Frontend (React) juntos, usamos o **Docker Compose**. Ele sobe toda a estrutura com um comando: `docker-compose up`.

````python
# Esboço do arquivo docker-compose.yml
'''
services:
  backend:
    build: ./backend
    ports: ["8000:8000"]
    env_file: [.env]
    
  frontend:
    build: ./frontend
    ports: ["5173:5173"]
'''
````
