## 1. A Teoria: Engenharia antes do Código

Muitos desenvolvedores cometem o erro de abrir o editor e sair digitando código sem um plano. Nós usaremos uma abordagem muito mais profissional: o **Spec-Driven Development (SDD)**. Isso significa que primeiro nós definimos o contrato de como as coisas vão funcionar, e só depois criamos o código que cumpre esse contrato.

Além disso, nosso projeto será guiado pelos princípios do **12-Factor App**, uma cartilha criada por engenheiros de ponta para garantir que aplicativos sejam robustos e fáceis de manter. Neste capítulo, aplicaremos os dois primeiros fatores:

1. **Base de Código Única:** Todo o nosso projeto ficará em um único repositório Git.
2. **Isolamento de Dependências:** Não vamos instalar bibliotecas soltas no seu computador. Criaremos uma "bolha" (Ambiente Virtual) exclusiva para este projeto.

---

## 2. A Prática Guiada: O Isolamento Tático (Setup)

Vamos preparar o seu equipamento. Abra o terminal integrado do VS Code na pasta raiz do seu projeto e siga os passos abaixo.

### Passo 1: Criando o Ambiente Virtual
O Python permite criar um ambiente isolado (`venv`) para que as bibliotecas deste projeto não entrem em conflito com outros projetos seus.

**Execute este comando no seu terminal:**
```bash
python -m venv venv
```

### Passo 2: Ativando o Ambiente Virtual
Agora precisamos "entrar" nessa bolha que acabamos de criar. O comando muda dependendo do seu sistema operacional.

**Execute no terminal (Escolha o do seu Sistema):**
```bash
# Para Windows:
.\venv\Scripts\activate

# Para Linux ou Mac:
source venv/bin/activate
```
*(Sucesso: Você verá um `(venv)` escrito no início da linha do seu terminal).* 

### Passo 3: Instalando as Dependências (NÃO PULE ESTE PASSO!)
Com a sua "bolha" (ambiente virtual) ativa, agora vamos trazer as ferramentas. O arquivo `requirements.txt` que criamos contém a lista de tudo o que o seu projeto precisa.

**Execute no terminal:**
```bash
cd backend
pip install -r requirements.txt
```

## 4. Troubleshooting (Resolução de Panes Comuns)

Mesmo com um planejamento rigoroso, imprevistos acontecem. Aqui estão as soluções para os problemas mais comuns nesta fase:

*   **Pane 1: O terminal diz que `python` não é reconhecido.**
    *   *Solução:* O Python não foi adicionado ao PATH do seu sistema. Reinstale o Python e marque a caixa "Add Python to PATH".
*   **Pane 2: Erro de "execução de scripts desabilitada" no Windows.**
    *   *Solução:* Abra o PowerShell como Admin e rode: `Set-ExecutionPolicy Unrestricted -Force`.
*   **Pane 3: `ModuleNotFoundError` mesmo após o pip install.**
    *   *Solução:* O VS Code está usando o Kernel errado. Clique em "Select Kernel" no topo direito e escolha o caminho do seu `venv`.
