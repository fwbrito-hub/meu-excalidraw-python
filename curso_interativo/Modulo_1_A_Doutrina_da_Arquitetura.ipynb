## 1. A Estratégia: Engenharia antes do Primeiro `import` 

No desenvolvimento de sistemas complexos, usamos o **Spec-Driven Development (SDD)**. Antes de escrever uma única linha de código, definimos o contrato: como os dados entram, como são processados e como saem. 

### Os Pilares do Sucesso (12-Factor App)
Nosso projeto segue os princípios das maiores empresas de tecnologia do mundo (como Heroku e Google). Neste setup inicial, aplicamos os dois primeiros fatores cruciais:

1.  **Codebase Única (Fator 1):** Um único repositório Git para toda a operação. Isso garante rastreabilidade e controle total sobre cada mudança.
2.  **Isolamento de Dependências (Fator 2):** Nunca instale bibliotecas globalmente no seu sistema. Criamos um "Bunker de Desenvolvimento" (Ambiente Virtual) onde as ferramentas deste projeto estão protegidas de conflitos externos.

---

## 2. A Prática: Construindo seu Bunker de Desenvolvimento

Vamos preparar seu equipamento de combate. Siga estes passos com precisão militar no terminal do VS Code.

### Passo 1: O Isolamento Tático (`venv`)
O Python permite criar o `venv` (Virtual Environment). Pense nele como uma bolha de proteção para seu projeto.

**Comando de Criação:**
```bash
python -m venv venv
```

### Passo 2: Ativação do Setor de Operações
Para que o computador saiba que deve usar a bolha que criamos, precisamos ativá-la.

**Comandos de Ativação:**
```powershell
# No Windows (PowerShell):
.\venv\Scripts\Activate.ps1

# No Linux/Mac (Bash):
source venv/bin/activate
```
*(Dica de Sucesso: Você verá o prefixo `(venv)` no seu terminal. Isso indica que você está operando dentro do bunker).* 

### Passo 3: Abastecimento de Ferramentas (`pip`)
Com o bunker ativo, instalaremos as armas pesadas: FastAPI, SQLAlchemy e o motor de IA Agno.

**Comando de Instalação:**
```bash
cd backend
pip install -r requirements.txt
```

## 4. Troubleshooting: O Manual de Campo do Desenvolvedor

Nenhum plano sobrevive ao primeiro contato com a realidade. Se algo deu errado, consulte estas soluções rápidas:

**Pane 1: `Scripts de execução desabilitados` no Windows.**
*   **Diagnóstico:** Política de segurança do Windows bloqueando o `activate`.
*   **Solução:** Abra o PowerShell como Administrador e rode: `Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser`.

**Pane 2: O VS Code não encontra as bibliotecas (sublinhado vermelho).**
*   **Diagnóstico:** O editor está olhando para o Python global, não para o seu bunker.
*   **Solução:** Pressione `Ctrl + Shift + P`, digite "Python: Select Interpreter" e escolha o que está dentro da pasta `venv`.

**Pane 3: `pip install` falhou com erro de conexão.**
*   **Diagnóstico:** Problemas de rede ou proxy.
*   **Solução:** Tente rodar `pip install --upgrade pip` primeiro e depois repita a instalação das ferramentas.
