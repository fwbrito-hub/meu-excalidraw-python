## 1. A Teoria: Agentes vs. Chatbots

Diferente de um chatbot comum que apenas conversa, um **Agente de IA** possui:
*   **Raciocínio:** Ele analisa o contexto (nossos elementos do Excalidraw).
*   **Memória:** Ele lembra do histórico da conversa.
*   **Ferramentas:** Ele pode consultar bancos de dados ou APIs externas.

No nosso caso, o Agente lerá a lista de elementos JSON (quadrados, setas, textos) e usará um modelo de linguagem (LLM) para interpretar a arquitetura do sistema que você desenhou.

## 4. Troubleshooting (Falhas na Matrix)

**Pane 1: `AuthenticationError` ou chave inválida.**
*   **Solução:** Verifique se o seu arquivo `.env` está na pasta correta (`backend/`) e se a chave não possui espaços extras.

**Pane 2: A IA responde coisas genéricas demais.**
*   **Solução:** Ajuste as `instructions` do Agente no arquivo `bot.py`. Quanto mais específico for o "Role" (Papel) do agente, melhor será a resposta.
