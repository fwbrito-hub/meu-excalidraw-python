# 📘 Capítulo 3: Persistência de Dados (Salvando os Desenhos)

**O Desafio:** Imagine desenhar uma arquitetura complexa por 2 horas e, ao reiniciar o servidor para uma atualização, tudo sumir. Sem persistência, seu software é apenas um castelo de areia na maré alta.

Neste módulo, vamos dar **memória permanente** ao nosso motor (API). Vamos configurar o banco de dados SQLite e usar o SQLAlchemy para que nossa aplicação consiga gravar e recuperar arquivos JSON de forma profissional.

---

## 1. A Teoria: O que é um ORM e por que SQLite?

No desenvolvimento moderno de alto nível, raramente escrevemos comandos SQL puros (como `INSERT INTO...`). Em vez disso, utilizamos um **ORM (Object-Relational Mapping)**.

### O Tradutor Universal (SQLAlchemy)
O ORM funciona como um tradutor: você lida com objetos Python (classes) e ele se encarrega de "conversar" com o banco de dados. Isso traz:
*   **Segurança:** Proteção nativa contra ataques de SQL Injection.
*   **Velocidade de Desenvolvimento:** Você foca na regra de negócio, não na sintaxe do banco.
*   **Portabilidade:** Hoje usamos SQLite, mas se amanhã precisarmos de um PostgreSQL (nuvem), mudamos apenas uma linha de configuração.

### O Banco Local (SQLite)
Seguindo o **Princípio 4 do 12-Factor (Backing Services)**, tratamos o banco como um recurso plugável. O SQLite é perfeito para nossa fase de prototipagem pois ele é apenas um arquivo dentro da sua pasta, sem necessidade de configurar servidores complexos como Docker ou instâncias pagas na nuvem.

## 2. A Prática Guiada: Modelando a Base de Dados

Vamos estruturar como a informação será guardada. Imagine uma ficha de cadastro para cada desenho. No nosso projeto, essa estrutura está no arquivo `backend/app/database.py`.

### A Anatomia do Modelo (`ProjetoDB`)
Nossa tabela `projetos` terá quatro colunas fundamentais:
1.  **ID:** Uma identidade única (UUID) para nunca confundirmos um desenho com outro.
2.  **Nome do Projeto:** Para organização do usuário.
3.  **Elementos JSON:** O coração do Excalidraw. Guardamos a lista de formas, cores e textos como um bloco de texto longo.
4.  **Data de Criação:** Registro temporal para auditoria e histórico.

---

## 3. Visualizando o Fluxo de Dados (A Jornada do Dado)

Para um arquiteto de software, entender o caminho é tão importante quanto o destino:
1.  **Frontend:** O usuário clica em "Salvar".
2.  **API (FastAPI):** Recebe o JSON e valida se os campos estão corretos.
3.  **ORM (SQLAlchemy):** Transforma o objeto Python em uma instrução SQL.
4.  **Disco Rígido:** O SQLite escreve os bits no arquivo `excalidraw_data.db`.

## 5. Troubleshooting (Resolução de Panes de Combate)

> **Pane 1: `sqlite3.OperationalError: database is locked`**
> *Causa:* Dois processos tentando escrever simultaneamente ou o arquivo está aberto em um visualizador externo (como DB Browser).
> *Solução:* Feche qualquer software de visualização de banco de dados antes de rodar a API.

> **Pane 2: Caracteres especiais (acentos) estranhos.**
> *Causa:* Problema de codificação.
> *Solução:* Garanta que seu editor (VS Code) está salvando os arquivos com codificação **UTF-8**.
