# 📘 Capítulo 3: Persistência de Dados (Salvando no Banco)

**Objetivo:** Garantir que o trabalho feito no frontend não se perca quando o servidor reiniciar. Vamos salvar os desenhos em um banco de dados SQLite.

---

## 1. A Teoria: Bancos de Dados e ORMs

Um servidor sem banco de dados tem "amnésia". Tudo o que ele recebe fica na memória RAM e some quando ele desliga. 

*   **SQLite:** Um banco de dados leve que mora em um arquivo `.db` na sua pasta.
*   **SQLAlchemy:** Uma ferramenta que nos permite tratar tabelas do banco de dados como classes do Python.

---

## 2. A Prática Guiada: A Estrutura de Dados

Já criamos o arquivo `backend/app/database.py`. Ele contém a "receita" da nossa tabela `projetos` conforme definimos no nosso **SDD**.

### Passo 1: Criando o Banco de Dados
Agora, vamos atualizar o `main.py` para que ele crie o arquivo do banco de dados assim que o servidor subir.

### Passo 2: Testando a Persistência
1. Abra o terminal onde o `uvicorn` está rodando.
2. O servidor deve reiniciar automaticamente.
3. Note que um novo arquivo chamado `projeto.db` aparecerá na sua pasta `backend`!

---

## 3. Troubleshooting (Resolução de Panes)

*   **Pane: `OperationalError: no such table: projetos`.**
    *   *Solução:* O comando de criar tabelas não foi executado. Verifique se o `database.criar_tabelas()` está no `main.py`.
