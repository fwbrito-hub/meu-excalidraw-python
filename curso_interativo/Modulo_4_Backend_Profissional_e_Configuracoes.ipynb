## 🛑 Lição 1: O Porteiro da API (CORS)

Por padrão, navegadores bloqueiam requisições vindas de domínios diferentes por segurança. 
*   **Frontend:** `http://localhost:5173` (Vite/React)
*   **Backend:** `http://localhost:8000` (FastAPI)

Sem o CORS configurado, o React **jamais** conseguirá salvar um desenho no seu banco. Vamos liberar o acesso de forma controlada.

```python
from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware

app = FastAPI()

# 🛡️ DEFINIÇÃO DOS DOMÍNIOS PERMITIDOS
origins = [
    "http://localhost:5173", # Porta padrão do Vite (React)
    "http://127.0.0.1:5173",
    "*", # PERIGO: O asterisco libera para QUALQUER site. Use apenas em desenvolvimento!
]

# ADICIONANDO O MIDDLEWARE DE CORS
app.add_middleware(
    CORSMiddleware,
    allow_origins=origins,           # Quais sites podem acessar?
    allow_credentials=True,          # Permite cookies/autenticação?
    allow_methods=["*"],             # Quais métodos (GET, POST, etc)?
    allow_headers=["*"],             # Quais headers?
)

@app.get("/")
def read_root():
    return {"mensagem": "CORS configurado e pronto para o React!"}
```

## 🔑 Lição 2: Pydantic Settings & Arquivo .env

Colocar a URL do banco de dados ou sua chave da Google AI no código é um **erro fatal** em produção. 
Usaremos o `pydantic-settings` para carregar essas variáveis automaticamente de um arquivo oculto chamado `.env`.

```python
from pydantic_settings import BaseSettings, SettingsConfigDict

class Settings(BaseSettings):
    # 📜 Definimos as variáveis e seus tipos
    DATABASE_URL: str = "sqlite:///./meu_projeto.db"
    GOOGLE_API_KEY: str
    APP_NAME: str = "Excalidraw IA"
    
    # ⚙️ Configuração para ler o arquivo .env
    model_config = SettingsConfigDict(env_file=".env")

# Instanciando as configurações
try:
    settings = Settings()
    print(f"Conectado ao Banco: {settings.DATABASE_URL}")
    print(f"App: {settings.APP_NAME}")
except Exception as e:
    print("❌ Erro: Você esqueceu de criar o arquivo .env ou faltam variáveis!")
```

## 🚜 Lição 3: Alembic - O Trator das Mudanças

No Módulo 2, usamos `Base.metadata.create_all()`. Isso é ótimo para testes, mas e se você quiser adicionar uma coluna "Cor do Projeto" depois que já tem 1000 desenhos salvos? O `create_all` não faz isso. O **Alembic** sim.

### Fluxo de Combate Alembic:
1. **`alembic init alembic`**: Cria a pasta de migrations.
2. **`alembic revision --autogenerate -m "Criar tabela projetos"`**: O Alembic olha suas classes do SQLAlchemy e gera o código de criação automaticamente.
3. **`alembic upgrade head`**: Aplica as mudanças no banco real.
