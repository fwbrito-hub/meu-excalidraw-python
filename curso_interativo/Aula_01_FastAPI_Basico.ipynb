### 🛠️ Passo 1: Importando as Dependências
Primeiro, vamos importar o FastAPI. Certifique-se de que seu ambiente virtual (`venv`) está ativado no VS Code e que você selecionou o Kernel correto para este Notebook.

````python
from fastapi import FastAPI
import uvicorn

# Instanciando a aplicação
app = FastAPI(title="Meu Curso de Excalidraw")

print("✅ FastAPI instanciado com sucesso!")
````

### 🌐 Passo 2: Criando a nossa primeira Rota
No FastAPI, definimos rotas usando decoradores como `@app.get("/")`. Isso diz ao servidor: "Quando alguém acessar o endereço principal, execute esta função".

````python
@app.get("/")
async def home():
    return {
        "status": "Online",
        "mensagem": "Bem-vindo à API do Excalidraw Customizado!",
        "aula": "01 - FastAPI Básico"
    }
````

### 🚀 Passo 3: Rodando o Servidor
