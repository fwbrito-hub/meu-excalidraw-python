## 🖌️ Lição 2: Injetando o Excalidraw

O Excalidraw fornece um pacote NPM oficial. Ele é pesado, mas extremamente poderoso. Vamos importá-lo como um componente comum do React.

````python
// Imagine este código dentro do seu App.tsx
import React, { useState } from "react";
import { Excalidraw } from "@excalidraw/excalidraw";

function App() {
  const [excalidrawAPI, setExcalidrawAPI] = useState(null);

  return (
    <div style={{ height: "100vh", width: "100vw" }}>
      <h1 style={{ textAlign: "center" }}>Meu Excalidraw Inteligente</h1>
      
      <Excalidraw 
        excalidrawAPI={(api) => setExcalidrawAPI(api)}
      />
    </div>
  );
}

export default App;
````

## 📡 Lição 3: Conectando com o Backend (O Botão Salvar)

Lembra da rota `POST /api/projetos/salvar` que criamos no FastAPI? Agora vamos usar a `fetch` API do navegador para enviar nossos dados para lá.

````python
const salvarDesenho = async () => {
  if (!excalidrawAPI) return;

  // 1. Extraímos os elementos e o estado do quadro
  const elementos = excalidrawAPI.getSceneElements();
  
  const payload = {
    nome_projeto: "Meu Diagrama Incrível",
    elementos_json: elementos
  };

  // 2. Enviamos para o FastAPI (que deve estar rodando na porta 8000)
  try {
    const response = await fetch("http://localhost:8000/api/projetos/salvar", {
      method: "POST",
      headers: { "Content-Type": "application/json" },
      body: JSON.stringify(payload),
    });
    
    const data = await response.json();
    alert("Projeto salvo com ID: " + data.id);
  } catch (error) {
    console.error("Erro ao salvar:", error);
    alert("Verifique se o Backend está ligado (e o CORS configurado)!");
  }
};
````
