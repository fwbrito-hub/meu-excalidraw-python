# ✅ Solução: Lição 03 - Segurança SaaS (Auth)

A forma profissional de proteger uma rota no FastAPI é injetando a dependência do usuário logado. No nosso projeto, essa função é a `get_current_user` do `app.main` (ou de um módulo especializado como `auth.py`).

```python
from fastapi import Depends
from app.main import get_current_user # Resposta do exercicio

@app.delete("/api/projetos/{id}")
async def deletar_projeto(id: int, user = Depends(get_current_user)):
    return {"msg": f"Projeto {id} deletado por {user.username}"}
```
